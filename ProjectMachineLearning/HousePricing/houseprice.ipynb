```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df.price,df.bedrooms, label = 'bedrooms')
plt.scatter(df.price,df.bathrooms, label = 'bathrooms')
plt.xlabel("Price ($)")
plt.xticks(rotation = 45)
plt.title("Price Between Bathrooms and Bedrooms")
plt.legend(loc = 'upper right')
plt.show()
```

```python
plt.scatter(df.price, df.grade, label = "Grade")
plt.scatter(df.price, df.condition, label = "Condition")
plt.scatter(df.price, df.waterfront, label = "Waterfront")
plt.scatter(df.price, df.view, label = "View")
plt.xticks(rotation = 45)
plt.legend(loc = "upper right")
plt.title("Price VS Quality House")
plt.xlabel("Price ($)")
plt.show()
```

```python
df.corr().style.background_gradient(cmap ="coolwarm")
```

```python
Y = df.price.values
X = df.drop(['price','id','date'], axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 10)
```

```python
print ('Data Train =', X_train.shape[0] ,  'Data Test =' , X_test.shape[0])
```

```python
from sklearn import linear_model
```

```python
regr = linear_model.LinearRegression()
regr.fit(X_train, Y_train)
```

```python
Y_predict = regr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(Y_predict,Y_test)
```
